```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_relevant.drop(columns='country', inplace=True)
df_relevant
```

```python
df_relevant.drop(columns='gender', inplace=True)
df_relevant
```

```python
sns.heatmap(df_relevant.corr(), annot=True, cmap='Reds')
```

```python
X = df_relevant.drop(columns='car purchase amount')
y = pd.DataFrame(df_relevant['car purchase amount'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = scaler.fit_transform(X)
y = scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=41)
```

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()

linear.fit(X_train, y_train)
linear.score(X_train, y_train)
```

```python
linear.coef_
```

```python
y_pred = linear.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"mae : {mae:.10f}")
print(f"mse : {mse:.10f}")
print(f"rmse : {rmse:.10f}")
print(f"r2 score : {r2:.10f}")
```

```python
plt.scatter(y_test, y_pred, alpha=0.7, color="blue")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2) # Diagonal line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted')
plt.show()
```
